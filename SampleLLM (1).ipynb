```python
import tensorflow as tf
import numpy as np

# tEXT --> DataSet
text = "hello world"
chars = sorted(list(set(text)))
char_to_ind = {u:i for i, u in enumerate(chars)}
ind_to_char = np.array(chars)

# Convert text to integers
input_sequence = [char_to_ind[c] for c in "hello"] # Input: "hello"
input_tensor = tf.convert_to_tensor([input_sequence])

#EMBEDDING
vocab_size = len(chars)
embedding_dim = 10
embedding_layer = tf.keras.layers.Embedding(vocab_size, embedding_dim)
dense_layer = tf.keras.layers.Dense(vocab_size)


embeddings = embedding_layer(input_tensor) # Shape: (1, 5, 10)
logits = dense_layer(embeddings)           # Shape: (1, 5, vocab_size)

last_token_logits = logits[0, -1, :]
probabilities = tf.nn.softmax(last_token_logits)

print(f"Input: 'hello'")
print(f"Model prediction shape: {logits.shape}")0
print(f"Most likely next character index: {np.argmax(probabilities)}")
print(f"Character: {ind_to_char[np.argmax(probabilities)]}")
```
